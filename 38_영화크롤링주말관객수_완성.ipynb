```python
import requests
import pandas as pd


key = '인증키' 
targetDt = '20220828'
weekGb = 1

url = 'http://kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchWeeklyBoxOfficeList.json?key={}&targetDt={}&weekGb={}'.format(key, targetDt, weekGb)

response = requests.get(url)
r_data = response.json()
weeklyBoxOfficeList = pd.DataFrame(r_data['boxOfficeResult']['weeklyBoxOfficeList'])  

weeklyBoxOfficeList = weeklyBoxOfficeList[['movieNm','openDt','rank','audiCnt','audiAcc','scrnCnt','showCnt']]
weeklyBoxOfficeList.index=weeklyBoxOfficeList.index+1

weeklyBoxOfficeList.rename(columns={'movieNm':'영화제목','openDt':'개봉일','rank':'당일 순위',
'audiCnt':'해당일 관객수','audiAcc':'누적관객수','scrnCnt':'해당일 상영 스크린수','showCnt':'해당일 상영 횟수'},inplace=True)


weeklyBoxOfficeList
```
