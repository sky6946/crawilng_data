```python
import requests
from bs4 import BeautifulSoup as bs

url = 'https://comic.naver.com/webtoon/weekday'
re = requests.get(url)
re.raise_for_status()

soup = bs(re.text,'html.parser')
```

```python
cartoons = soup.find_all('a',attrs={'class':'title'})

for cartoon in cartoons:
    print(cartoon.get_text())
```
